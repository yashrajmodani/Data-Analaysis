```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10)
y = np.arange(1,19)
```

```python
y = y[:9]
```

```python
y
```

```python
#Create a line plot

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot Title Here')
plt.scatter(x,y, s= 50) #Increase marker sie to 50
plt.show()
```
