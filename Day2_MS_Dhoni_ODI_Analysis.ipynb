```python
import pandas as pd
import datetime
import numpy as np

import seaborn as sns
```

```python
df = pd.read_csv("MS_Dhoni_ODI_record.csv")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
# replace v in opposition  [Data cleaning ]

# df['opposition'] = df['opposition'].str.replace('v','',regex=False) # method1
df['opposition'] =df['opposition'].apply(lambda x: x[2:])  # method2
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst= True)
df['Year'] = df['date'].dt.year.astype(int)

df
```

```python
df.Year  # or df['Year']
```

```python
df.info()
```

```python
df
```

```python
# batting avg = total runs scored /(Innigs count - not outs count)

df['score'] = df['score'].apply(str)
df['not_out'] = np.where(df['score'].str.endswith('*'),1,0)
df
```

```python
df.drop(columns='odi_number',inplace=True)
```

```python
df
```

```python
df_new = df.loc[((df['score']!='DNB') & (df['score'] != "TDNB")),'runs_scored':]
df_new
```

```python
# fixing the data types of numerical columns
df_new.info()
df_new['runs_scored'] =df_new['runs_scored'].astype(int)
df_new['balls_faced'] =df_new['balls_faced'].astype(int)
df_new['strike_rate'] =df_new['strike_rate'].astype(float)
df_new['fours'] =df_new['fours'].astype(int)
df_new['sixes'] =df_new['sixes'].astype(int)
```

```python
df_new.info()
```

```python
# career states

first_match_date = df['date'].dt.date.min().strftime('%B %d, %Y')
print(first_match_date)

last_match_date = df['date'].dt.date.max().strftime('%B %d, %Y')
print(last_match_date)
```

```python
numOfmatches = df.shape[0]
print(f"Number of matches : {numOfmatches}")

numOfinnings = df_new.shape[0]
print(f"Number of innings : {numOfinnings}")

not_outs = df_new['not_out'].sum()
print(f"Number of innings : {not_outs}")

runs_scored = df_new['runs_scored'].sum()
print(f"runs_scored : {runs_scored}")

balls_faced   = df_new['balls_faced'].sum()
print(f"balls_faced   : {balls_faced}")
```

```python
# batting avg = total runs scored /(Innigs count - not outs count)
career_sk = runs_scored / balls_faced
print("Career strike: {:.2f}".format(career_sk))

career_avg = runs_scored /(numOfinnings - not_outs )
print("Career avg: {:.2f}".format(career_avg))
```

```python
# Calculate strike rate per opposition and show as a heatmap

#Group by opposition
grouped_by_opposition = df_filtered.groupby('opposition')
grouped_by_opposition
```

```python
#Aggregating "Balls face" and 'Runs_scored using sum

agg_sum = grouped_by_opposition.agg({'balls_faced':'sum','runs_scored':'sum'})
agg_sum
```

```python
# Assigning the result to df_strike_rate

df_strike_rate = agg_sum
df_strike_rate
```

```python
#Calculate strike rate

df_strike_rate['strike_rate'] = (df_strike_rate['runs_scored']/df_strike_rate['balls_faced'])*100
df_strike_rate
```

```python
#Bad graph -  strike rate scale is different from runs scored and balls face

sns.heatmap(df_strike_rate,linewidth = 0.5,annot = True, fmt = ".0f")
plt.title("Strike rate per opposition")
plt.xticks(rotation = 69)
plt.show()
```

```python
df2 = df_new.groupby('opposition')[['runs_scored', 'sixes', 'fours']].sum().reset_index()
df2['runs_in_boundries'] = (df2['fours'] * 4) + (df2['sixes'] * 6)
df2
```

```python
df2 = df2[df2['runs_scored'] > 200]
df2
```

```python
plt.bar(df2['opposition'], df2['runs_in_boundries'])
plt.xticks(rotation=45)
plt.show()
```

```python
#Stacked bar plot
plt.figure(figsize=(10,6))
plt.bar(df2['opposition'], df2['runs_scored'],bottom = df2['runs_in_boundries'], label='Runs Scored')
plt.bar(df2['opposition'], df2['runs_in_boundries'], label='Runs In Boundries')
plt.xlabel('Oppostion')
plt.ylabel('Runs')
plt.title('Runs scored  and runs in boundries by opposition')
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
