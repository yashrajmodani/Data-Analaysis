19. Write a Pandas program to replace the missing values with the most frequent
values present in each column of a given DataFrame

```python
import pandas as pd

# Sample data as dictionary
data = {
    'ord_no': [70001.0, None, 70002.0, 70004.0, None, 70005.0, None, 70010.0, 70003.0, 70012.0, None, 70013.0],
    'purch_amt': [150.50, None, 65.26, 110.50, 948.50, None, 5760.00, 1983.43, None, 250.45, 75.29, 3045.60],
    'sale_amt': [10.50, 20.65, None, 11.50, 98.50, None, 57.00, 19.43, None, 25.45, 75.29, 35.60],
    'ord_date': ['2012-10-05', '2012-09-10', None, '2012-08-17', '2012-09-10', '2012-07-27', '2012-09-10', '2012-10-10', '2012-10-10', '2012-06-27', '2012-08-17', '2012-04-25'],
    'customer_id': [3002, 3001, 3001, 3003, 3002, 3001, 3001, 3004, 3003, 3002, 3001, 3001],
    'salesman_id': [5002.0, 5003.0, 5001.0, None, 5002.0, 5001.0, 5001.0, None, 5003.0, 5002.0, 5003.0, None]
}

# Create DataFrame
df = pd.DataFrame(data)

# Replace missing values with the most frequent value in each column
df = df.apply(lambda x: x.fillna(x.mode()[0]) if x.isnull().sum() > 0 else x)

# Output the result
print(df)
```

# 20. Write a Pandas program to create a hitmap for more information about thedistribution of missing values in a given DataFrame.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'ord_no': [70001.0, None, 70002.0, 70004.0, None, 70005.0, None, 70010.0, 70003.0, 70012.0, None, 70013.0],
    'purch_amt': [150.50, None, 65.26, 110.50, 948.50, None, 5760.00, 1983.43, None, 250.45, 75.29, 3045.60],
    'sale_amt': [10.50, 20.65, None, 11.50, 98.50, None, 57.00, 19.43, None, 25.45, 75.29, 35.60],
    'ord_date': ['2012-10-05', '2012-09-10', None, '2012-08-17', '2012-09-10', '2012-07-27', '2012-09-10', '2012-10-10', '2012-10-10', '2012-06-27', '2012-08-17', '2012-04-25'],
    'customer_id': [3002, 3001, 3001, 3003, 3002, 3001, 3001, 3004, 3003, 3002, 3001, 3001],
    'salesman_id': [5002.0, 5003.0, 5001.0, None, 5002.0, 5001.0, 5001.0, None, 5003.0, 5002.0, 5003.0, None]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of Missing Values in DataFrame', fontsize=16)
plt.show()
```
