# ***`1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates. `***

# ***`17. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.`***

```python
# Define the date range
start_date = pd.to_datetime('2020-04-01')
end_date = pd.to_datetime('2020-09-30')

# Filter the data
df['Date'] = pd.to_datetime(df['Date'])
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
filtered_df = df.loc[mask]

# Calculate daily percentage returns
filtered_df.set_index('Date', inplace=True)
daily_returns = filtered_df['Adj Close'].pct_change()

# Plot the daily percentage returns
daily_returns.plot(figsize=(10, 7), legend=True, linestyle='--', marker='o')
plt.title('Daily % Return of Alphabet Inc. Stock Price\n01-04-2020 to 30-09-2020')
plt.grid(True)
plt.show()
```

# ***`18. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates. `***

```python
# Define the date range
start_date = pd.to_datetime('2020-04-01')
end_date = pd.to_datetime('2020-09-30')

# Filter the data
df['Date'] = pd.to_datetime(df['Date'])
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
filtered_df = df.loc[mask]

# Calculate daily returns and rolling standard deviation (volatility)
filtered_df.set_index('Date', inplace=True)
daily_returns = filtered_df['Close'].pct_change()
volatility = daily_returns.rolling(window=30).std()

# Plot the volatility
volatility.plot(figsize=(10, 7))
plt.title('Volatility Over Time of Alphabet Inc. Stock Price\n01-04-2020 to 30-09-2020')
plt.grid(True)
plt.show()
```

# ***`19. Write a Pandas program to create a histogram to visualize daily return distribution of Alphabet Inc. stock price between two specific dates.`***

```python
# Define the date range
start_date = pd.to_datetime('2020-04-01')
end_date = pd.to_datetime('2020-09-30')

# Filter the data
df['Date'] = pd.to_datetime(df['Date'])
mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
filtered_df = df.loc[mask]

# Calculate daily percentage returns
filtered_df.set_index('Date', inplace=True)
daily_returns = filtered_df['Adj Close'].pct_change()

# Plot the histogram
sns.histplot(daily_returns.dropna(), bins=100, color='purple')
plt.title('Daily Return Distribution of Alphabet Inc. Stock Price\n01-04-2020 to 30-09-2020')
plt.grid(True)
plt.show()
```
