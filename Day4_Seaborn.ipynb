```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Generate random data

np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
#Create a scatter plot with matplotlib

plt.figure(figsize = (10,6))
plt.scatter(x,y,color = 'blue',alpha = 0.6)
plt.title("Scatter plot using matplotlib",fontsize = 15)
plt.xlabel("X-axis",fontsize = 12)
plt.ylabel("Y-axis",fontsize = 12)
plt.grid(True)
plt.show()
```

```python
#Create a scatter plot with seaborn

sns.set(style = 'darkgrid')    # Set style with white grid background
plt.figure(figsize = (10,6))
sns.scatterplot(x = x, y = y)
plt.title("Scatter plot using Seaborn",fontsize = 15)
plt.xlabel("X-axis",fontsize = 12)
plt.ylabel("Y-axis",fontsize = 12)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/heart.csv")
df
```

```python
#Simple Pandas

ages.plot.line('age','chol');
plt.show()
```

```python
#Matplotlib
fig,ax = plt.subplots()
ax.plot(ages['age'],ages['chol'],ls = ":",lw = 1.7)
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.set_title("Line Plot")
plt.show()
```

```python
#Seaborn

sns.lineplot(x = 'age',y = 'chol',data = ages,linestyle = "-",linewidth = 1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Line Plot")
plt.show()
```

# ***DM_office_sales_Dataset***

#Count plot :
A simple plot, it merely shows the total count of rows per category

A histogram across a  categorical, instead of qantitive  , variable

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("/content/dm_office_sales.csv")
df
```

```python
plt.figure(figsize =(10,4),dpi = 200) #dpi = Dot per inch of space
sns.countplot(x = 'division',data = df)
plt.show()
```

```python
plt.figure(figsize =(10,4),dpi = 200) #dpi = Dot per inch of space
sns.countplot(x = 'level of education',data = df)
plt.show()
```

```python
#Breakdown within another category with 'hue'


#hue is used to create a category or breakdown into a category

plt.figure(figsize =(10,4),dpi = 200) #dpi = Dot per inch of space
sns.countplot(x = 'level of education',hue = 'training level',data = df)
plt.show()
```

```python
#Using palette

plt.figure(figsize =(10,4),dpi = 200) #dpi = Dot per inch of space
sns.countplot(x = 'level of education',hue = 'training level',data = df,palette = "Set1")
plt.show()
```

```python

plt.figure(figsize =(10,4),dpi = 200) #dpi = Dot per inch of space

#Paired would be a good choice if there was a distinct jump from 0 and 1 to2 and 3
sns.countplot(x = 'level of education',hue = 'training level',data = df,palette = "Paired")
plt.show()
```
