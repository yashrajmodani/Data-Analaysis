```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew
```

```python
#Generate random data with a larger sample size
sample_size = 100000
ratings = np.random.randint(1,6,sample_size)
ratings
```

```python
#Modify data for right skew

right_skew_data = np.concatenate((ratings,np.random.randint(1,2,sample_size//2)))
right_skew_data.shape
```

```python
#Modify data for left skew

left_skew_data = np.concatenate((ratings,np.random.randint(4,6,sample_size//2)))
left_skew_data.shape
```

```python
#Function to calculate mean, median and skewness

def calculate_stats(data):
  mean = np.mean(data)
  median = np.median(data)
  skewness = skew(data)
  return mean, median, skewness
```

```python
# Function to plot histogram with mean and median lines

def plot_histogram(ax,data,title):
  ax.hist(data,bins = np.arange(1,7) - 0.5,rwidth = 0.8,alpha = 0.7 )
  ax.set_title(title)
  ax.set_xlabel("rating")
  ax.set_ylabel("frequency")
  ax.set_xticks(range(1,6))
  ax.grid(axis = 'y',alpha = 0.5)

  # calculate mean, median and skewness

  mean,median,skewness = calculate_stats(data)
  ax.axvline(x = mean, color = 'r',ls = '--',label = f'Mean:{mean:.2f}')
  ax.axvline(x = median, color = 'b',ls = '-.',label = f'Median:{median}')
  ax.legend()



  #Print mean,median, skewness

  print(f"{title}:")
  print(f'Mean:{mean:.2f}')
  print(f'Median:{median:}')
  print(f'Skewness:{skewness:.2f}')
  print()


```

```python
fig, axs = plt.subplots(1,3,figsize = (15,5))

plot_histogram(axs[0],ratings,'No Skew')
plot_histogram(axs[1],right_skew_data,'Right Skew')
plot_histogram(axs[2],left_skew_data,'Left Skew')

plt.tight_layout()
plt.show()
```

**Functionality on Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
airline_df = pd.read_csv('/content/airline_dec_2008_50k.csv')
airline_df
```

```python
airline_df.columns
```

```python
distance = airline_df['Distance']
distance
```

```python
mean_distance = distance.mean()
median_distance = distance.median()
skewness_distance = distance.skew()
```

```python
plt.figure(figsize = (8,6))
sns.histplot(distance , kde = True, color = 'c',bins = 30)
plt.axvline(mean_distance,color = 'r',ls ='--',label = f'Mean:{mean_distance:.2f}')
plt.axvline(median_distance,color = 'black',ls ='--',label = f'Median:{median_distance:.2f}')

plt.title(f'Distribution of Distance\n skewness: {skewness_distance:.2f}')
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```
