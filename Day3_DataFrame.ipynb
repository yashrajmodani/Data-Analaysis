# **Air Bnb Dataset**

```python
#another index example - air BNB dataset from kaggle
#default index

df_airbnb = pd.read_csv("AB_NYC_2019.csv")
df_airbnb.head(3)
```

```python
#Make id column the index
df2 = df_airbnb.set_index("id")
df2.head(3)
df2.name[3647] #Get the name for id 3647
```

```python
#set pandas display options to show all columns
numeric_columns = df2.select_dtypes(include = ['float64','int64'])

#Include 'room_type' column in numeric_column DataFrame
numeric_columns['room_type'] = df2['room_type']
df3 = numeric_columns.groupby('room_type').mean()
pd.set_option('display.max_columns',None)
df3.head(3)
```

```python
df3.index
```

```python
#Note that room_type has become an index - we can again convert it into a normal column
df3 = df3.reset_index()
pd.reset_option('display.max_columns') #Also reset the column display to default
df3
```

```python
#Rows
#Get a single row
#Integer based
df3.iloc[0]
```

```python
#Name based
#First set the index to the column which will be used in locating  the row

df = df_airbnb.set_index('host_name')
df.loc['John']
```

```python
#Remove a row
df = df.drop('John',axis = 0)
df.head()
```

```python
#Multiple conditions
df_new = df[(df['total_bill'] > 30) & (df['gender'] == 'Male')]
df_new.head()
```

```python
df_new = df[(df['total_bill'] > 30) & ~(df['gender'] == 'Male')]
df_new.head()
```

```python
df_new = df[(df['total_bill'] > 30) & (df['gender'] != 'Male')]
df_new.head()
```

```python
#Creating groups on multiple columns

df_grouped = df.groupby(['gender','day'])
df_grouped['tip'].mean()
```

```python
## Agg Function

#apply multiple aggregation function to different columns when grouping data
#calculate both the mean and sum of the 'tip' column for each combination of 'Gender' and 'day'

#Group the dataframe by both 'gender' and 'day' and calculate the mean and yhe sum of 'tip' for each combination

df_grouped = df.groupby(['gender','day']).agg({'tip': ['mean','sum']}).reset_index()
df_grouped
```

```python
#Rename the columns for clarity

df_grouped.columns = ['gender','day','Average tip','total_tip']
df_grouped
```
