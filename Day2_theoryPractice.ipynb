```python
import pandas as pd
import numpy as np
import datetime
```

```python
#DAta cleaning - opposition name says'v Aus' etc, we can remove 'v '

df['opposition'] = df['opposition'].apply(lambda x: x[2:]) #method2
# df['opposition'] = df['opposition'].str.replace('v ','',regex = False) #method1

#regex = False means that the first string ('v ') is not a regex , but a literal string
```

```python
df
```

```python
#Add a 'feature' - 'year' column using the match date column
#First convert date column into datetime format

df['date'] = pd.to_datetime(df['date'], dayfirst = True)
df['year'] = df['date'].dt.year.astype(int)
df
```

```python
df.year  # or df['Year']
```

```python
df.info()
```

```python
df
```

```python
#Batting average = (Total runs scored) / (innings count - not out counts)
#creae a column to distinguish between out and not out
df['score'] = df['score'].apply(str)
df['not_out'] = np.where(df['score'].str.endswith('*'),1,0)
df
```

```python
#dropping the odi number feature because it adds no value to the analysis
df.drop(columns = 'odi_number',inplace = True)
```

```python
df
```

```python
df_new = df.loc[((df['score']!='DNB') & (df['score'] != "TDNB")),'runs_scored':]
df_new
```

```python
# fixing the data types of numerical columns
df_new.info()
df_new['runs_scored'] =df_new['runs_scored'].astype(int)
df_new['balls_faced'] =df_new['balls_faced'].astype(int)
df_new['strike_rate'] =df_new['strike_rate'].astype(float)
df_new['fours'] =df_new['fours'].astype(int)
df_new['sixes'] =df_new['sixes'].astype(int)
```

```python
df_new.info()
```

```python
#Career stats
first_match_date = df['date'].dt.date.min().strftime('%B,%d,%Y') #first match
print('First_match: ', first_match_date)

last_match_date = df['date'].dt.date.max().strftime('%B,%d,%Y')  #Last match
print('Last_match: ', last_match_date)
```

```python
number_of_matches = df.shape[0] #number of matches played in career
print('Number of matchses played:' , number_of_matches)

number_of_inns = df_new.shape[0] #number of innings
print(f"number of innings: {number_of_inns}")

not_outs = df_new['not_out'].sum() # number of not out in career
print(f"not outs: {not_outs}")

runs_scored = df_new['runs_scored'].sum() # runs scored in career
print(f"runs scored: {runs_scored}")

balls_faced = df_new['balls_faced'].sum() #balls faced
print(f"Balls faced in the creer: {balls_faced}")
```

```python
# batting avg = total runs scored /(Innigs count - not outs count)
career_sr = (runs_scored / balls_faced)*100 #career strike rate
print(f"career sr: {career_sr}")
career_sk = runs_scored / balls_faced
print("Career strike: {:.2f}".format(career_sk))

career_avg = (runs_scored/(number_of_inns - not_outs))#career  avg
print(f"Career avg: {career_avg}")
```

```python
hundreds = (df_new['runs_scored'] >= 100).sum()
print(f"Number of 100s: {hundreds}")

fifties = (df_new['runs_scored'] >= 50).sum()
print(f"Number of 50s: {fifties}")
```

```python
fours = df_new['fours'].sum(0) #number of fours in career
print(f"Number of 4s: {fours}")

sixes = df_new['sixes'].sum() #number of sixes in career
print(f"Number of 6s: {sixes}")
```
